```python
def agecat(data):
    if data <20 :                 return 1
    elif data>=20 and data <25:   return 2
    elif data>=25 and data <30:   return 3
    elif data>=30 and data <35:   return 4
    elif data>=35 and data <40:   return 5
    elif data>=40 and data <45:   return 6
    elif data>=45 and data <50:   return 7
    elif data>=50 and data <55:   return 8
    elif data>=55 and data <60:   return 9
    elif data>=60             :   return 10
    else :                        return np.nan

n["AgeGrp"]=n.Age.apply(agecat)
n
```

### Q.6 The Marketing manager wants to understand the total amount spent on various products so that we can find what percentage of the amount is spent on which product

```python
n
```

```python
n.columns
```

```python
n["TotalMnt"]=n["MntWines"]+n["MntFruits"]+n["MntMeatProducts"]+n["MntFishProducts"]+n["MntSweetProducts"]+n["MntGoldProds"]
n.head()
```

```python
col=[("MntWines","wine%"),("MntFruits","fruit%"),("MntMeatProducts","meat%"),("MntFishProducts","fish%"),("MntSweetProducts","sweet%"),("MntGoldProds","gold%")]
for i,j in col:
    n[j]=(n[i]/n["TotalMnt"])*100
n
```

```python
col=[("MntWines","wine"),("MntFruits","fruit"),("MntMeatProducts","meat"),("MntFishProducts","fish"),("MntSweetProducts","sweet"),("MntGoldProds","gold")]
for i,j in col:
    ans=((n[i].sum()/n["TotalMnt"].sum())*100).round()
    print(f"percentage {j} purchased= {ans} %")
```

### Ques 13. Convert the Year_Birth column into categorical column. Create categories according to the analysis done so far.

```python
sns.histplot(n.Year_Birth, bins=15, kde=1)
plt.xticks(rotation=90)
plt.grid()
plt.show()
```

```python
def BirthYearCat(data):
    if data < 1955: return "baby boomers"
    elif data >=1955 & data < 1970: return "Generation Jones"
    elif data>=1970 and data <1985:   return "Generation X"
    elif data>=1985 and data <2000:   return" Millennials"
    else :                        return np.nan

n["BirthYearCat"]=n["Year_Birth"].apply(BirthYearCat)
n
```
